```python
##Linear Regression Line
from matplotlib import pyplot as plt

x = [1, 2, 4]
y = [2, 1, 3]

plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.axis([0, 10, 0, 10])
plt.scatter(x, y)

for i in range(0, len(x)):
    label = "(" + str(x[i]) + ", " + str(x[i]) + ")"
    plt.annotate(label, xy = (x[i], y[i]), xytext = (5, -5), textcoords = 'offset points')

x_avg = 0
for i in range(0, len(x)):
    x_avg += x[i] 
x_avg = x_avg / len(x)

y_avg = 0
for i in range(0, len(x)):
    y_avg += y[i]
y_avg = y_avg / len(y)

xs_avg = 0
for i in range(0, len(x)):
    xs_avg += x[i] * x[i]
xs_avg = xs_avg / len(x)

xy_avg = 0
for i in range(0, len(x)):
    xy_avg += x[i] * y[i]
xy_avg = xy_avg / len(x)

m = (x_avg * y_avg - xy_avg) / (x_avg * x_avg - xs_avg)
b = y_avg - m * x_avg

print(m)
print(b)

x4line = []
y4line = []
for i in range(0, 11):
    x4line.append(i)
    y4line.append(m*i + b)
    
plt.plot(x4line, y4line, 'k-', label = 'y='+str(m)+'x+'+str(b))
plt.legend(loc=9)

plt.show()
```

```python
##Line chart
from matplotlib import pyplot as plt

x = [1, 2, 4, 8, 16, 32, 64, 128, 256]
y = []
for i in range(len(x)-1, -1, -1):
    y.append(x[i])
z = []
for i in range(0, len(x)):
    z.append(x[i] + y[i])
    
xs = []
for i in range(0, len(x)):
    xs.append((i*1)*100)

plt.plot(xs, x, 'g-', label = 'variance')
plt.plot(xs, y, 'r-.', label = 'bias^2')
plt.plot(xs, z, 'b:', label = 'total error')

plt.legend(loc=0)

plt.show()
```
